```python
import pickle
import plotly.graph_objects as go
import helpers
```

```python
full_links_dict = pickle.load(open('data/billionairesdict.pkl','rb'))
links_dict = full_links_dict# helpers.trim_dict(full_links_dict, 100)
# del links_dict['Donald Trump']
# del links_dict['Elon Musk']
```

```python
connectedness_total = [len(links_dict[name]['linkshere']) for name in links_dict.keys()]

views = [links_dict[name]['total_views'] for name in links_dict.keys()]
graph_data = go.Scatter(
    y=views, x=connectedness_total,
    mode='markers',
    marker_color='rgb(204,40,81)',
    marker_size=4,
    text=list(links_dict.keys()),
    hovertemplate =
    '<b>%{text}</b>'+
    '<br><b>Links</b>: %{x}'+
    '<br><b>Views</b>: %{y}<br><extra></extra>')
fig = go.Figure(data=[graph_data])

axis_config = dict(
    title_font=dict(
        size=24,
        family='Rubik',
        color='rgb(255,255,255)'),
    tickcolor='rgb(180,155,191)',
    tickfont=dict(
        family='Karla',
        color='rgb(255,255,255)',
        size=20),
    gridcolor='rgb(44,32,50)'
    )
y_axis = dict(title='Page views', type='log')
y_axis.update(axis_config)
x_axis = dict(title='Number of links to page')
x_axis.update(axis_config)
fig.update_layout(
    title='Number of links to page vs. page views over the last 60 days',
    title_font=dict(
        color='rgb(255,255,255)',
        size=32,
        family='Rubik'), 
    yaxis=y_axis,
    xaxis=x_axis,
    paper_bgcolor='rgb(22,16,25)',
    plot_bgcolor='rgb(22,16,25)',
    width=1920,
    height=1080)
fig.show()
```

```python
fig.write_html("plots/scatterplot.html")
```
